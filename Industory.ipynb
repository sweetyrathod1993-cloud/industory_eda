```python
# Count records by industry
industry_counts = data['industry'].value_counts()

# Pie chart
plt.figure(figsize=(4, 4))
plt.pie(
    industry_counts,
    labels=industry_counts.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Industry Distribution')
plt.axis('equal')  # Makes pie chart circular

plt.show()
```

### The industry distribution is perfectly balanced, with Technology, Retail, Manufacturing, Healthcare, and Finance each representing 20% of the total dataset.

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 12))

# 1. Industry - Count Plot
sns.countplot(
    data=data,
    y='industry',
    order=data['industry'].value_counts().index,
    color='steelblue',
    ax=axes[0, 0]
)
axes[0, 0].set_title('Industry Distribution')

# 2. Industry - Pie Chart
industry_counts = data['industry'].value_counts()
axes[0, 1].pie(
    industry_counts,
    labels=industry_counts.index,
    autopct='%1.1f%%',
    colors=plt.cm.Set3.colors
)
axes[0, 1].set_title('Industry Share')

# 3. Employee Count - Histogram
sns.histplot(
    data=data,
    x='employee_count',
    kde=True,    
    color='green',
    stat='percent',
    ax=axes[0, 2]
)
axes[0, 2].set_title('Employee Count Distribution')

# 4. Annual Revenue - Histogram
sns.histplot(
    data=data,
    x='annual_revenue_million',
    kde=True,
    color='orange',
    stat='percent',
    ax=axes[1, 0]
)
axes[1, 0].set_title('Annual Revenue Distribution')

# 5. Profit Margin - Violin Plot
sns.violinplot(
    x=data['profit_margin_percent'],
    ax=axes[1, 1],
    color='pink'
)
axes[1, 1].set_title('Profit Margin Distribution')

# 6. Market Rating - KDE Plot
sns.kdeplot(
    data=data,
    x='market_rating',
    fill=True,
    color='red',
    ax=axes[1, 2]
)
axes[1, 2].set_title('Market Rating Distribution')

plt.tight_layout()
plt.show()
```

### Bivariate analysis

```python
# Correlation Heatmap
plt.figure(figsize=(8, 6))

corr = data[
    ['employee_count',
     'annual_revenue_million',
     'profit_margin_percent',
     'customer_count',
     'market_rating']
].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='viridis',
    fmt='.2f',
    linewidths=1,
    square=True
)

plt.title('Correlation Matrix', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

### The heatmap shows the correlation between:

### Employee Count
### Annual Revenue
### Profit Margin
### Customer Count
### Market Rating

### All correlation coefficients are very close to 0
### Increasing employee count does not appear to increase or decrease revenue.

### Multivariate Analysis

```python
industry_summary = data.groupby('industry').agg({
    'annual_revenue_million': 'mean',
    'employee_count': 'mean'
}).reset_index()

plt.figure(figsize=(10,6))

plt.scatter(
    industry_summary['employee_count'],
    industry_summary['annual_revenue_million'],
    s=300
)

for i, row in industry_summary.iterrows():
    plt.text(
        row['employee_count'],
        row['annual_revenue_million'],
        row['industry']
    )

plt.xlabel('Average Employee Count')
plt.ylabel('Average Annual Revenue (Million USD)')
plt.title('Industry vs Revenue vs Employee Count')
plt.show()
```

### 1. Retail Industry Generates the Highest Average Revenue

### Retail has the highest average annual revenue at approximately $556 million.
### Its employee count is around 2,608, indicating strong revenue generation without having the largest workforce.

### 2. Finance Has the Largest Workforce

### Finance has the highest average employee count (~2,622 employees).
### It generates the second-highest revenue (~$553 million).
### This suggests that a larger workforce does not necessarily translate into the highest revenue.

### 3. Healthcare Shows the Lowest Revenue

### Healthcare has an average employee count similar to Technology and Retail (~2,610 employees).
### However, it records the lowest average revenue (~$543 million).
### This indicates lower revenue efficiency compared to other industries.

### 4. Manufacturing Has the Smallest Workforce

### Manufacturing has the lowest average employee count (~2,583 employees).
### It also has relatively low revenue (~$546 million).
### This may indicate smaller operational scale compared to other industries.

### 5. Weak Relationship Between Employee Count and Revenue

### Industries with similar employee counts show noticeable differences in revenue.
### For example:
### Retail and Healthcare have nearly the same employee count.
### Retail generates significantly higher revenue.

```python
country_summary = data.groupby('country').agg({
    'customer_count': 'mean',
    'annual_revenue_million': 'mean'
}).reset_index()

plt.figure(figsize=(10,6))

plt.scatter(
    country_summary['customer_count'],
    country_summary['annual_revenue_million'],
    s=300,
    alpha=0.7
)

for i, row in country_summary.iterrows():
    plt.text(
        row['customer_count'],
        row['annual_revenue_million'],
        row['country']
    )

plt.xlabel('Average Customer Count')
plt.ylabel('Average Annual Revenue (Million USD)')
plt.title('Country × Customer Count × Revenue')
plt.show()
```

### 1. USA Generates the Highest Revenue
### The USA has an average revenue of approximately $553 million, the highest among all countries.
### Its average customer count is around 50,200.
### This suggests strong revenue generation despite not having the largest customer base.

### 2. Canada Has the Largest Customer Base
### Canada has the highest average customer count (~51,450 customers).
### However, its average revenue (~$548 million) is lower than that of the USA and India.
### This indicates that a larger customer base does not necessarily result in higher revenue.

### 3. India Shows Strong Revenue Efficiency
### India has the lowest average customer count (~49,950 customers).
### Despite this, it generates the second-highest average revenue (~$552 million).
### This suggests higher revenue generated per customer compared to other countries.

### 4. Germany Records the Lowest Revenue
### Germany has a customer count similar to the USA.
### However, it has the lowest average revenue (~$546 million).
### This may indicate lower revenue generation efficiency.

### 1. No Clear Relationship Between Company Age and Revenue
### Revenue values are spread widely across all founding years (1990–2019).
### Both older and newer companies generate revenues ranging from approximately $50 million to $1,050 million.
### This suggests that company age does not strongly influence revenue.

### 2. Revenue Distribution Is Similar Across Industries
### Companies from Finance, Technology, Retail, Manufacturing, and Healthcare are scattered throughout the chart.
### No industry consistently occupies the highest or lowest revenue levels.
### Revenue patterns appear similar regardless of industry.

### 3. High-Revenue Companies Exist in All Time Periods
### Companies with revenues above $900 million are present across nearly all founding years.
### This indicates that both older and newer companies can achieve strong financial performance.

### 4. No Upward or Downward Trend
### If newer companies generated higher revenue, points would gradually rise toward recent years.
### If older companies generated higher revenue, points would be concentrated at higher levels in earlier years.
### Neither pattern is visible.

### 5. Even Distribution Suggests a Balanced Dataset
### The points form vertical bands with similar revenue ranges for each year.
### This indicates that revenue is distributed relatively uniformly across founding years and industries.
